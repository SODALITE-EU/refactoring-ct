# Benchmark Viewer

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

benchmark_rt = [0.47969902014430565, 0.28847068051497143, 0.13503361940383912, 0.37898067263669744, 0.3790503978729248, 0.24035538567437065, 0.15516410451946835, 0.2071336030960083, 0.19662977087086644]
benchmark_req = [79, 70, 20, 110, 60, 50, 50, 30, 29]

x_val = np.arange(len(benchmark_rt))
print("avg rt %s", benchmark_rt)
print("req %s", benchmark_req)
plt.plot(x_val, benchmark_rt, '--', label="AVG RT")
plt.show()
plt.plot(x_val, benchmark_req, label="#REQ")
plt.show()
```
